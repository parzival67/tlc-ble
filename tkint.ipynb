```python
import tkinter as tk
from PIL import Image, ImageTk

def display_png_image(file_path):
    root = tk.Tk()
    root.title("PNG Image Display")

    # Open the PNG image file using PIL (Pillow)
    image = Image.open(file_path)
    
    # Convert the PIL Image to a Tkinter PhotoImage
    tk_image = ImageTk.PhotoImage(image)

    # Create a Label widget and set the image
    label = tk.Label(root, image=tk_image)
    label.pack()

    # Start the Tkinter event loop
    root.mainloop()

# Example usage:
png_file_path = "part.png"
display_png_image(png_file_path)
```

```python
import tkinter as tk
from tkinter import ttk
from tkPDFViewer import tkPDFViewer as pdf 

def display_pdf(file_path, parent):
    # creating object of ShowPdf from tkPDFViewer.
    v1 = pdf.ShowPdf() 
  
    # Adding pdf location and width and height.
    v2 = v1.pdf_view(parent, pdf_location=file_path)
    v2.grid(row=0, column=0, padx=10, pady=10)  # Adjust the padding as needed

def create_table(parent):
    # Configure the style of Heading in Treeview widget
    style = ttk.Style()
    style.theme_use('clam')
    style.configure('Treeview.Heading', font=('JetBrains Mono', 9, 'bold'), background="black", foreground="white")

    # scrollbar
    scroll = tk.Scrollbar(parent)
    scroll.grid(row=0, column=2, sticky='ns')

    # Treeview widget
    tree = ttk.Treeview(parent, yscrollcommand=scroll.set)
    tree.tag_configure('1', background='lightgrey')
    tree.tag_configure('0', background='white')
    tree.grid(row=0, column=1, padx=10, pady=10, sticky='nsew')

    scroll.config(command=tree.yview)

    tree['columns'] = ('c1', 'c2', 'c3')

    # format our column
    tree.column("#0", width=0, stretch=tk.NO)
    tree.column("#1", anchor=tk.CENTER, width=50)
    tree.column("#2", anchor=tk.CENTER, width=130)
    tree.column("#3", anchor=tk.CENTER, width=130)

    # Create Headings 
    tree.heading("#0", text="", anchor=tk.CENTER)
    tree.heading("#1", text="ID", anchor=tk.CENTER)
    tree.heading("#2", text="DIMENSION", anchor=tk.CENTER)
    tree.heading("#3", text="MEASURED VALUE", anchor=tk.CENTER)

    # add data 
    tree.insert(parent='', index='end', iid=0, text='', values=('1', '10.23530', 'NIL'), tags=('0'))
    tree.insert(parent='', index='end', iid=1, text='', values=('2', '1.027475', 'NIL'), tags=('1'))
    tree.insert(parent='', index='end', iid=2, text='', values=('3', '203.4532', 'NIL'), tags=('0'))
    tree.insert(parent='', index='end', iid=3, text='', values=('4', '10.23530', 'NIL'), tags=('1'))
    tree.insert(parent='', index='end', iid=4, text='', values=('5', '1.027475', 'NIL'), tags=('0'))
    tree.insert(parent='', index='end', iid=5, text='', values=('6', '203.4532', 'NIL'), tags=('1'))
    tree.insert(parent='', index='end', iid=6, text='', values=('7', '10.23530', 'NIL'), tags=('0'))
    tree.insert(parent='', index='end', iid=7, text='', values=('8', '1.027475', 'NIL'), tags=('1'))
    tree.insert(parent='', index='end', iid=8, text='', values=('9', '203.4532', 'NIL'), tags=('0'))

# Main application
root = tk.Tk()
root.title("Image and Table Display")

# Create a frame to hold both the image and the table
frame = tk.Frame(root)
frame.pack()

# Display PDF on the left
display_pdf("part.pdf", frame)

# Create table on the right
create_table(frame)

# Configure row and column weights
root.grid_rowconfigure(0, weight=1)
root.grid_columnconfigure(1, weight=1)

root.mainloop()
```
